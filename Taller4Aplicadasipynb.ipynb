# 1) Serie Fibonacci:
Escribe un programa en Python que genere los primeros n números
de la serie de Fibonacci y visualízalos en un gráfico de líneas utilizando Matplotlib. El
usuario debe poder elegir el número n

```python
import matplotlib.pyplot as plt

def fibonacci(n):
    if n <= 0:
        return []
    elif n == 1:
        return [0]

    fib = [0, 1]
    for i in range(2, n):
        fib.append(fib[i-1] + fib[i-2])
    return fib

n = int(input("Ingrese el valor de n: "))
fib_series = fibonacci(n)

print(fib_series)

plt.plot(fib_series, marker='o', linestyle='-', color='b', label="Serie Fibonacci")
plt.xlabel("Índice")
plt.ylabel("Valor")
plt.title("Gráfico de la Serie de Fibonacci")
plt.legend()
plt.grid()
plt.show()
```

# 6) Pendiente de una recta:
Escribe un programa que tome dos puntos en el plano (x1, y1)
y (x2, y2), calcule la pendiente de la recta que los une, y grafique dicha recta

```python
import matplotlib.pyplot as plt

def calcular_pendiente(x1, y1, x2, y2):
    if x1 == x2:
        raise ValueError("La pendiente es indefinida (división por cero)")
    return (y2 - y1) / (x2 - x1)

def graficar_recta(x1, y1, x2, y2):
    pendiente = calcular_pendiente(x1, y1, x2, y2)

    # Crear puntos para la recta
    x_values = [x1, x2]
    y_values = [y1, y2]

    plt.plot(x_values, y_values, marker='o', linestyle='-', color='b', label=f'y = {pendiente:.2f}x + b')
    plt.scatter([x1, x2], [y1, y2], color='r', label='Puntos')

    plt.xlabel("X")
    plt.ylabel("Y")
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(True, linestyle='--', linewidth=0.5)
    plt.legend()
    plt.title(f"Recta entre dos puntos\nPendiente: {pendiente:.2f}")
    plt.show()

# Ejemplo de uso
x1 = float(input("Ingrese x1: "))
y1 = float(input("Ingrese y1: "))
x2 = float(input("Ingrese x2: "))
y2 = float(input("Ingrese y2: "))

graficar_recta(x1, y1, x2, y2)


```

# 7) Movimiento parabólico:
Escribe un programa que simule el movimiento de un
proyectil bajo la influencia de la gravedad (sin resistencia del aire).

 Grafica la trayectoria parabólica para diferentes ángulos de lanzamiento


```python
import numpy as np
import matplotlib.pyplot as plt

def simular_movimiento(v0, angulo, g=9.81):
    angulo_rad = np.radians(angulo)

    # Tiempo de vuelo total
    t_vuelo = (2 * v0 * np.sin(angulo_rad)) / g

    # Tiempo en pequeños intervalos
    t = np.linspace(0, t_vuelo, num=100)

    # Posiciones en X y Y
    x = v0 * np.cos(angulo_rad) * t
    y = v0 * np.sin(angulo_rad) * t - 0.5 * g * t**2

    return x, y

def graficar_trayectorias(v0, angulos):
    plt.figure(figsize=(10, 5))

    for angulo in angulos:
        x, y = simular_movimiento(v0, angulo)
        plt.plot(x, y, label=f'{angulo}°')

    plt.xlabel("Distancia (m)")
    plt.ylabel("Altura (m)")
    plt.title("Movimiento Parabólico de un Proyectil")
    plt.legend()
    plt.grid()
    plt.show()

# Parámetros
v0 = float(input("Ingrese la velocidad inicial (m/s): "))
angulos = list(map(float, input("Ingrese los ángulos de lanzamiento separados por espacio: ").split()))

graficar_trayectorias(v0, angulos)
```

# 9) Punto Extra:
Fractales con Python: Escribe un programa que genere el fractal de
Sierpinski utilizando ciclos (while), y visualiza el resultado con Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
import random

def sierpinski(iterations):
    points = np.array([[0, 0], [1, 0], [0.5, np.sqrt(3)/2]])  # Vértices del triángulo
    x, y = [0.5], [0.5]  # Punto inicial

    i = 0
    while i < iterations:
        vertex = random.choice(points)  # Elegir un vértice al azar
        x.append((x[-1] + vertex[0]) / 2)
        y.append((y[-1] + vertex[1]) / 2)
        i += 1

    plt.figure(figsize=(6, 6))
    plt.scatter(x, y, s=0.2, color="black")
    plt.axis("off")
    plt.title("Fractal de Sierpinski")
    plt.show()

# Ejecutar la función con 10000 iteraciones
sierpinski(10000)
```
